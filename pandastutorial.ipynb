```python
# Pearson Coefficient of Skewness
mean_value = df['Annual Salary'].mean()
median_value = df['Annual Salary'].median()
stdVal = df['Annual Salary'].std()
x =(3*(mean_value-median_value))/stdVal
print(x)
print('Proved that the data is possitively skewed')
```
